```python
import matplotlib.pyplot as plt
```

```python
df.hist(bins=50, figsize=(20,15))
```

```python
df_amount=df['amount']
df_amount.hist(bins=100, figsize=(50,15))
```

```python
df_count=df['count']
df_count.hist(bins=50, figsize=(50,15))
```

```python
import seaborn as sns
sns.distplot(df_amount, hist=False, kde=True, rug=False, color='red', vertical=True, axlabel="AMOUNT", label="SMUW")
```

```python
sns.jointplot(x="amount", y="count", data=df, kind="scatter")
```

```python
sns.jointplot(x="amount", y="count", data=df,  kind="kde", space=0, color="g").plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
sns.jointplot(x="amount", y="count", data=df, kind="kde", space=0, zorder=0, n_levels=6)
```

```python
Q1 = df['amount'].quantile(q=0.25)
Q3 = df['amount'].quantile(q=0.75)
IQR = Q3-Q1
IQR
```

```python
df_IQR=df[(df['amount']<Q3+IQR*1.5)& (df['amount']>Q1-IQR*1.5)] 
```

```python
df_IQR
```

```python
df.boxplot(column='count', return_type='both')
```

```python
df.boxplot(column='amount', return_type='both')
```

```python
ndf=df.filter(['amount', 'count'])
ndf=ndf[(ndf['count']<10) & (ndf['amount']<500000)]
ndf.hist(bins=50, figsize=(50,15))
```

```python
sns.jointplot(x="amount", y="count", data=ndf)
```

```python
sns.jointplot(x="amount", y="count", data=ndf, kind="kde", space=0, zorder=0, n_levels=6)
```

```python
import numpy as np
ndf['logamount']=np.log(ndf['amount'])
```

```python
sns.jointplot(x="logamount", y="count", data=ndf, kind="kde", space=0, zorder=0, n_levels=6)
```
